```python
# Import the modules 
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read and display the CSV with Pandas
covid_df = pd.read_csv("Resources/Covid19_Confirmed_dataset.csv")
covid_df
```

```python
#  Delete the useless columns
covid_df.drop(["Lat", "Long"], axis = 1, inplace = True)
covid_df
```

```python
# Aggregating the rows by the country
covid_df_aggregated = covid_df.groupby("Country/Region").sum()
covid_df_aggregated.drop(["Province/State"], axis = 1, inplace = True)
covid_df_aggregated
```

```python
# Visualizing data related to the Canada and China
covid_df_aggregated.loc["Canada"].plot()
covid_df_aggregated.loc["China"].plot()
plt.legend()
```

```python
# Spread of the virus in the first 5 days
covid_df_aggregated.loc["China"][ : 5].plot()
```

```python
# Caculating the first derivative of the curve
covid_df_aggregated.loc["China"].diff().plot()
```

```python
# Find maxmimum infection rate for China
covid_df_aggregated.loc["China"].diff().max()
```

```python
# Find maxmimum infection rate for the US
covid_df_aggregated.loc["US"].diff().max()
```

```python
# Find maxmimum infection rate for Canada
covid_df_aggregated.loc["Canada"].diff().max()
```

```python
# Find maximum infection rate for all of the countries
countries = list(covid_df_aggregated.index)

# Create an empty list
max_infection_rate = []

for i in countries:
    x = covid_df_aggregated.loc[i].diff().max()
    max_infection_rate.append(x)

covid_df_aggregated["max_infection_rate"] = max_infection_rate

covid_df_aggregated["max_infection_rate"]
```

```python
# Create a new dataframe with only needed column
covid_df_updated = pd.DataFrame(covid_df_aggregated["max_infection_rate"])
covid_df_updated
```

```python
# Importing the WorldHappinessReport.csv dataset
happiness_report = pd.read_csv("Resources/worldwide_happiness_report.csv")
happiness_report
```

```python
# Drop the useless columns
happiness_report.drop(["Overall rank", "Score", "Generosity", "Perceptions of corruption"], axis = 1, inplace = True)
happiness_report
```

```python
# Changing the indices of the dataframe
happiness_report.set_index("Country or region", inplace = True)
happiness_report
```

```python
# Join two dataset we have prepared
data = covid_df_updated.join(happiness_report, how = "inner")
data
```

```python
# Correlation matrix
data.corr()
```

```python
# Plotting GDP vs maximum Infection rate
data["log_max_infection_rate"] = np.log(data["max_infection_rate"])

sns.scatterplot(data = data, x = "GDP per capita", y = "log_max_infection_rate")
plt.title("GDP per capita vs Maximum Infection Rate")
plt.xlabel("GDP per capita")
plt.ylabel("Maximum Infection Rate")
plt.show()
```

```python
sns.regplot(data = data, x = "GDP per capita", y = "log_max_infection_rate")
```
